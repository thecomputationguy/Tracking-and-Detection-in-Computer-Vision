```python
def quaternionMetric(q_a, q_plus):
    theta = 2 * np.arccos((np.dot(q_a, q_plus) / (np.linalg.norm(q_a) * np.linalg.norm(q_plus))))
    return theta
```

```python
# To extract the quaternions and store them in a np array for ease of calculations

def arrayToSeq(dataAsList): 
    dataAsArray = np.array([])
    for i in range(1,5):
        dataAsArray = np.append(dataAsArray, dataAsList[i])
    return dataAsArray
```

```python
# To extract the class of the image in question

def findObjectType(filename):
    types = ["ape", "benchvise", "cam", "cat", "duck"]
    objectType = "something"
    
    for i, name in enumerate(types):
        if name in filename:
            objectType = name
        else:
            continue
            
    return objectType                
```

```python
def anchor(testData):
    anchorFile = random.choice(testData)    
    return anchorFile
```

```python
def puller(anchorImage, templateData):
    similarity = 1000 
    
    anchorImageType = findObjectType(anchorImage[0])
    
    anchorQuaternion = arrayToSeq(anchorImage)
    queryQuaternion = np.array([1.0, 2.0, 3.0, 4.0])    
    
    for index, file in enumerate(templateData):        
        queryImageType = findObjectType(file[0])
        
        if anchorImageType == queryImageType:
            template = file
            queryQuaternion = arrayToSeq(template)
            metric = quaternionMetric(anchorQuaternion, queryQuaternion)        
        
            if metric < similarity:
                similarity = metric
                imageName = file[0]
                pullerIndex = index
        else:
            continue
        
    return imageName        
```

```python
def pusher(pullerImage, templateData):
    
    while(True):
        pusherImage = random.choice(templateData)
        if pusherImage[0] == pullerImage[0]:
            continue
        else:
            break
            
    return pusherImage[0]
```

```python
def batchGenerator(testData, databaseData, batchSize = 20):
    dataBatch = []
    
    for i in range(0, batchSize):
        anchorImage = anchor(testData)
        pullerImageName = puller(anchorImage, databaseData)
        pusherImageName = pusher(pullerImageName, databaseData)
        
        image = cv2.imread(anchorImage[0], cv2.IMREAD_COLOR)
        normalizedImage = cv2.normalize(image, None, alpha=-1, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
        dataBatch.append(normalizedImage)
        
        image = cv2.imread(pullerImageName, cv2.IMREAD_COLOR)
        normalizedImage = cv2.normalize(image, None, alpha=-1, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
        dataBatch.append(normalizedImage)
        
        image = cv2.imread(pusherImageName, cv2.IMREAD_COLOR)
        normalizedImage = cv2.normalize(image, None, alpha=-1, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
        dataBatch.append(normalizedImage)
        
        print("Batches Generated : ", i + 1)
        
    np.save("dataBatch", dataBatch)
    print("File Saved!")
    return dataBatch
```

```python
def DBBatchGenerator(databaseData):
    temp = []
    DBdata = []
    batchSize = databaseData.size
    
    for i in range(0, batchSize):
        DBImage = databaseData[i][0]
        DBImageType = findObjectType(DBImage)
        
        image = cv2.imread(DBImage, cv2.IMREAD_COLOR)
        normalizedImage = cv2.normalize(image, None, alpha=-1, beta=1, norm_type=cv2.NORM_MINMAX, dtype=cv2.CV_32F)
        temp.append([normalizedImage, DBImageType]) 
        
        DBdata.append(temp)
        
        print("Images Read : ", i + 1, " | type : ", DBImageType)
        
    np.save("database_images_data", DBdata)
    print("File Saved!")
    return DBdata
```
